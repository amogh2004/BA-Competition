```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.max_columns', None)
sba_loans = pd.read_csv('SBAnational.csv')
sba_loans
```

```python
# Convert 'ApprovalDate' and 'DisbursementDate' to datetime format
sba_loans['ApprovalDate'] = pd.to_datetime(sba_loans['ApprovalDate'], errors='coerce')
sba_loans['DisbursementDate'] = pd.to_datetime(sba_loans['DisbursementDate'], errors='coerce')

# Calculate the number of days between ApprovalDate and DisbursementDate
sba_loans['DisbursementTime'] = (sba_loans['DisbursementDate'] - sba_loans['ApprovalDate']).dt.days

# Convert 'ApprovalDate' and 'DisbursementDate' to datetime format
sba_loans['ApprovalDate'] = pd.to_datetime(sba_loans['ApprovalDate'], errors='coerce')
sba_loans['ChgOffDate'] = pd.to_datetime(sba_loans['ChgOffDate'], errors='coerce')

# Calculate the number of days between ApprovalDate and DisbursementDate
sba_loans['ChgOffTime'] = (sba_loans['ChgOffDate'] - sba_loans['ApprovalDate']).dt.days

# Calculate the number of days between DisbursementDate and ChgOffDate
sba_loans['DisbursementToChgOffTime'] = (sba_loans['ChgOffDate'] - sba_loans['DisbursementDate']).dt.days
```

```python
sba_loans = sba_loans.drop(columns=['LoanNr_ChkDgt', 'Name', 'City', 'Zip', 'Bank', 'ApprovalDate', 'ChgOffDate', 'DisbursementDate'])
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to each specified column
sba_loans['State'] = label_encoder.fit_transform(sba_loans['State'])
sba_loans['BankState'] = label_encoder.fit_transform(sba_loans['BankState'])
sba_loans['NAICS'] = label_encoder.fit_transform(sba_loans['NAICS'])
```

```python
# Remove non-numeric characters and convert the column to integer
sba_loans['ApprovalFY'] = sba_loans['ApprovalFY'].astype(str)  # Convert all entries to string to handle mixed types
sba_loans['ApprovalFY'] = sba_loans['ApprovalFY'].str.extract('(\d{4})')[0]  # Extract only the year part
sba_loans['ApprovalFY'] = pd.to_numeric(sba_loans['ApprovalFY'], errors='coerce')  # Convert to numeric, coerce errors to NaN

# Convert to integer data type
sba_loans['ApprovalFY'] = sba_loans['ApprovalFY'].astype(int)
```

```python
# Drop rows where NewExist has values other than 1 or 2
sba_loans = sba_loans[sba_loans['NewExist'].isin([1, 2])]

# Replace values in the NewExist column
sba_loans['NewExist'] = sba_loans['NewExist'].replace({1: 0, 2: 1})

# Ensure the column is of integer type
sba_loans['NewExist'] = sba_loans['NewExist'].astype('Int64')  # Using 'Int64' to allow NaN values in integer columns
```

```python
# Drop rows where RevLineCr has values other than 'Y' or 'N'
sba_loans = sba_loans[sba_loans['RevLineCr'].isin(['Y', 'N'])]

# Convert 'Y' to 1 and 'N' to 0 in the RevLineCr column
sba_loans['RevLineCr'] = sba_loans['RevLineCr'].replace({'Y': 1, 'N': 0})
```

```python
# Drop rows where LowDoc has values other than 'Y' or 'N'
sba_loans = sba_loans[sba_loans['LowDoc'].isin(['Y', 'N'])]

# Convert 'Y' to 1 and 'N' to 0 in the LowDoc column
sba_loans['LowDoc'] = sba_loans['LowDoc'].replace({'Y': 1, 'N': 0})
```

```python
# Replace values in FranchiseCode: 1 or 0 should become 0, everything else should become 1
sba_loans['FranchiseCode'] = sba_loans['FranchiseCode'].replace({1: 0, 0: 0}).apply(lambda x: 1 if x != 0 else 0)
```

```python
# List of columns to convert
currency_columns = ['DisbursementGross', 'BalanceGross', 'ChgOffPrinGr', 'GrAppv', 'SBA_Appv']

# Remove dollar signs and commas, then convert to float
for col in currency_columns:
    sba_loans[col] = sba_loans[col].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Encode MIS_Status: 'CHGOFF' as 1, 'P I F' as 0
sba_loans['MIS_Status'] = sba_loans['MIS_Status'].replace({'CHGOFF': 1, 'P I F': 0})

# Convert MIS_Status to integer type with NaN values allowed
sba_loans['MIS_Status'] = sba_loans['MIS_Status'].astype('Int64')

# Convert DisbursementTime to integer type with NaN values allowed
sba_loans['DisbursementTime'] = sba_loans['DisbursementTime'].astype('Int64')

# Convert ChgOffTime to integer type with NaN values allowed
sba_loans['ChgOffTime'] = sba_loans['ChgOffTime'].astype('Int64')
```

```python
from scipy.stats import ttest_ind

# Define the list of numerical columns
numerical_columns = [
    'Term', 'NoEmp', 'CreateJob', 'RetainedJob', 'DisbursementGross', 
    'BalanceGross', 'ChgOffPrinGr', 'GrAppv', 'SBA_Appv', 
    'DisbursementTime', 'ChgOffTime', 'MIS_Status'
]

# Drop rows with any NaN values in the numerical columns or target column 'MIS_Status'
sba_loans_cleaned = sba_loans[numerical_columns].dropna()
```

```python
sba_loans_cleaned = sba_loans_cleaned.astype('int64')

sba_loans_cleaned.info()
```

```python
# Separate data based on the target variable
high_risk = sba_loans_cleaned[sba_loans_cleaned['MIS_Status'] == 1]
low_risk = sba_loans_cleaned[sba_loans_cleaned['MIS_Status'] == 0]

# Perform T-test for each numerical feature
ttest_results = {}
for column in numerical_columns:
    t_stat, p_value = ttest_ind(high_risk[column], low_risk[column])
    ttest_results[column] = p_value

# Display significant features (p-value < 0.05)
significant_features = {k: v for k, v in ttest_results.items() if v < 0.05}
significant_features
```

```python
# Calculate correlation between numerical features and the target variable
correlations = sba_loans_num.corrwith(sba_loans['MIS_Status']).abs()
print("Correlation with MIS_Status:\n", correlations)
```

### Summary of Significant Correlations:
1. Moderate Correlations: ChgOffPrinGr, Term have moderate correlations with MIS_Status, indicating they may be important features for modeling.
2. Weak Correlations: Features like SBA_Appv, GrAppv, DisbursementGross, and DisbursementTime have weak correlations but might still be useful.
3. Very Weak or No Correlation: Features like NoEmp, CreateJob, RetainedJob, BalanceGross, and ChgOffTime are weakly correlated and may have minimal impact on predicting MIS_Status.

### Conclusion
1. Key Predictors: ChgOffPrinGr and Term appear to be the most relevant features based on correlation.
2. Less Relevant Predictors: Features with very low correlation may not add much predictive value and could potentially be excluded or further examined for non-linear relationships in advanced modeling steps.

```python
from scipy.stats import chi2_contingency

# List of categorical columns to test
categorical_columns = ['NewExist', 'FranchiseCode', 'UrbanRural', 'RevLineCr', 'LowDoc']  

# Perform Chi-Square test for each categorical feature
chi2_results = {}
for column in categorical_columns:
    contingency_table = pd.crosstab(sba_loans[column], sba_loans['MIS_Status'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    chi2_results[column] = p

# Display significant features (p-value < 0.05)
significant_features = {k: v for k, v in chi2_results.items() if v < 0.05}
print("Significant categorical features based on Chi-Square test (p < 0.05):\n", significant_features)
```

```python
from scipy.stats import f_oneway

# List of ordinal categorical columns (if any)
ordinal_columns = ['State', 'BankState', 'NAICS', 'ApprovalFY'] 

anova_results = {}
for column in ordinal_columns:
    groups = [sba_loans[sba_loans[column] == category]['MIS_Status'].dropna() for category in sba_loans[column].unique()]
    f_stat, p_value = f_oneway(*groups)
    anova_results[column] = p_value

# Display significant features (p-value < 0.05)
significant_features_anova = {k: v for k, v in anova_results.items() if v < 0.05}
print("Significant ordinal categorical features based on ANOVA test (p < 0.05):\n", significant_features_anova)
```

```python
from scipy.stats import chi2_contingency, f_oneway
import pandas as pd

# Example categorical columns, modify with actual dataset's categorical column names
categorical_columns = ['NewExist', 'FranchiseCode', 'UrbanRural', 'RevLineCr', 'LowDoc']
ordinal_columns = ['State', 'BankState', 'NAICS', 'ApprovalFY']  

# Chi-Square Test for all categorical features
chi2_results = {}
for column in categorical_columns:
    # Generate contingency table for chi-square test
    contingency_table = pd.crosstab(sba_loans[column], sba_loans['MIS_Status'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    chi2_results[column] = {'Chi2 Statistic': chi2, 'p-value': p, 'Degrees of Freedom': dof}

# Display Chi-Square results
chi2_results_df = pd.DataFrame(chi2_results).T  # Transpose for readability

# ANOVA for ordinal categorical features (if any exist)
anova_results = {}
for column in ordinal_columns:
    groups = [sba_loans[sba_loans[column] == category]['MIS_Status'].dropna() for category in sba_loans[column].unique()]
    f_stat, p_value = f_oneway(*groups)
    anova_results[column] = {'F-statistic': f_stat, 'p-value': p_value}

# Display ANOVA results
anova_results_df = pd.DataFrame(anova_results).T  # Transpose for readability
```
